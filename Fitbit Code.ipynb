```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

filepath = "/Users/qixinzhao/Desktop/Fitbit_Data.xlsx"

df = pd.read_excel(filepath)

print(df.columns)
print(df.head())
```

```python
# Remove empty lines
df["Sleep duration (hours)"] = pd.to_numeric(df["Sleep duration (hours)"], errors='coerce')
# Check data types (make sure sleep, steps, temperature, etc. are numeric)
df["Sleep duration (hours)"] = pd.to_numeric(df["Sleep duration (hours)"], errors='coerce')
df["Steps(Fitbit)"] = pd.to_numeric(df["Steps(Fitbit)"], errors='coerce')
df["Mood Score (1-5)"] = pd.to_numeric(df["Mood Score (1-5)"], errors='coerce')
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Hours of Sleep vs Fitbit Steps
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df,
                x="Sleep duration (hours)",
                y="Steps(Fitbit)",
                hue="Weather (Sunny/Cloudy/Rainy)",
                palette="viridis")
plt.title("Sleep Duration vs. Fitbit Steps by Weather")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Steps (Fitbit)")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_rainy = df[df["Weather (Sunny/Cloudy/Rainy)"] == "Rainy"]
df_sunny = df[df["Weather (Sunny/Cloudy/Rainy)"] == "Sunny"]
df_cloudy = df[df["Weather (Sunny/Cloudy/Rainy)"] == "Cloudy"]

sns.set(style="whitegrid")

plt.figure(figsize=(6, 4))
sns.scatterplot(data=df_rainy, x="Sleep duration (hours)", y="Steps(Fitbit)", color="blue")
plt.title("Rainy Days: Sleep Duration vs Fitbit Steps44")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Steps (Fitbit)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.scatterplot(data=df_sunny, x="Sleep duration (hours)", y="Steps(Fitbit)", color="orange")
plt.title("Sunny Days: Sleep Duration vs Fitbit Steps")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Steps (Fitbit)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.scatterplot(data=df_cloudy, x="Sleep duration (hours)", y="Steps(Fitbit)", color="gray")
plt.title("Cloudy Days: Sleep Duration vs Fitbit Steps")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Steps (Fitbit)")
plt.tight_layout()
plt.show()
```

This graph demonstrates the relationship between sleep duration and daily steps under different weather conditions. Overall, there is no obvious linear correlation trend between the number of steps and sleep duration, but in terms of weather categorization, the number of steps is generally higher under sunny days, suggesting that good weather may promote higher levels of physical activity, while under cloudy and rainy days, the distribution of the number of steps is relatively low and the activity is reduced. From this, it can be tentatively hypothesized that weather, as an environmental factor, may influence daily physical activity levels to some extent.

```python
# Average Steps vs Weather Type
df["Weather (Sunny/Cloudy/Rainy)"] = df["Weather (Sunny/Cloudy/Rainy)"].str.strip().str.capitalize()
plt.figure(figsize=(7, 5))
sns.barplot(
    data=df,
    x="Weather (Sunny/Cloudy/Rainy)",
    y="Steps(Fitbit)",
    errorbar=None,
    palette="pastel"
)
plt.title("Average Fitbit Steps by Weather Type")
plt.xlabel("Weather Type")
plt.ylabel("Average Steps")
plt.tight_layout()
plt.show()
```

This bar graph shows the average Fitbit steps in different weather types. As you can see from the graph, the average number of steps is highest in cloudy weather, followed by sunny ， while the lowest number of steps is recorded during rainy weather. This suggests that there may be some influence of weather conditions on daily physical activity levels, especially on rainy days when people's activity levels may be significantly suppressed. It is worth noting, however, that the highest step counts on cloudy days may also be influenced by the interaction of other factors (e.g., temperature or mood), and further analysis of the control variables is needed subsequently before verifying the significance of this trend.

```python
# Mood Score vs Steps
plt.figure(figsize=(7, 5))
sns.boxplot(data=df,
            x="Mood Score (1-5)",
            y="Steps(Fitbit)",
            palette="Set2")
plt.title("Mood Score vs. Fitbit Steps")
plt.xlabel("Mood Score (1-5)")
plt.ylabel("Steps (Fitbit)")
plt.tight_layout()
plt.show()
```

This bar graph shows the average Fitbit steps in different weather types. As you can see from the graph, the average number of steps is highest in cloudy weather, followed by sunny days, while the lowest number of steps is recorded during rainy days. This suggests that there may be some influence of weather conditions on daily physical activity levels, especially on rainy days when people's activity levels may be significantly suppressed. It is worth noting, however, that the highest step counts on cloudy days may also be influenced by the interaction of other factors (e.g., temperature or mood), and further analysis of the control variables is needed subsequently before verifying the significance of this trend.

```python
# Fitbit vs Manual Estimated Steps
plt.figure(figsize=(8, 5))
sns.regplot(data=df,
            x="Manual Estimation of Steps",
            y="Steps(Fitbit)",
            scatter_kws={"color": "blue"},
            line_kws={"color": "red"})
plt.title("Manual vs. Fitbit Steps")
plt.xlabel("Manual Step Estimate")
plt.ylabel("Fitbit Steps")
plt.tight_layout()
plt.show()

import scipy.stats as stats
spearman_corr, p_value = stats.spearmanr(df["Manual Estimation of Steps"], df["Steps(Fitbit)"])
print(f"Spearman correlation: {spearman_corr:.2f}, p-value: {p_value:.4f}")
```

This graph shows the relationship between manually estimated steps and the number of steps actually recorded by the Fitbit. The graph shows a strong positive correlation between the two, with the data points roughly distributed along a regression straight line, indicating a high degree of consistency between the manually estimated steps and the values recorded by the device. This indicates that although there is some error in manual estimation, it can still reflect the actual physical activity level more accurately overall, and thus the manual estimation has some reference value when the data is missing or the device is unavailable.

```python
corr = df[["Steps(Fitbit)", "Sleep duration (hours)", "Mood Score (1-5)", "Average Temperature (℃)"]].corr()
print("correlation matrix：")
print(corr)
```

The correlation matrix shows a strong correlation between sleep duration and Fitbit steps (r = 0.52), suggesting that sleep duration may have a positive effect on physical activity levels the next day. The correlation between mood scores and step count was weak (r = 0.085), and mean temperature was almost independent of step count (r ≈ 0). There was also some correlation between sleep duration and mood scores (r = 0.48), suggesting that better sleep may lead to a more positive mood state.
